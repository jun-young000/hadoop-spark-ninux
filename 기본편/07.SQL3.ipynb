```python
spark.sql("""
SELECT LTRIM('    HELLO    '), RTRIM('    HELLO    '), TRIM('    HELLO    '),
LPAD('HELLO', 10, '*'), RPAD('HELLO', 10, '*')
""").show()
```

```python
ltrim(안부)

지정된 문자열 값에 대해 왼쪽 끝에서 공백을 자릅니다.
```

```python
trim(안부)

지정된 문자열 열의 양쪽 끝에서 공백을 자릅니다.
```

```python
rtrim(안부)

지정된 문자열 값에 대해 오른쪽 끝에서 공백을 자릅니다.
```

```python
lpad(col, len, 패드)

len 의 길이만큼 오른쪽 정렬해서 채우고  왼쪽부터 공백 나머지를 패드모양으로 채워라
```

```python
rpad(col, len, 패드)

마우스 오른쪽 패드 문자열 열은 폭 렌 과 패드 .
```

```python
# 그렇다면 예를 들면 성명 세글자가 있다면 가운데 글자만 *로  하려면 
# 어떻게 해야 할까.
```

```python
from pyspark.sql.functions import ltrim, rtrim, trim, lpad, rpad
```

```python
retails.select(ltrim(lit("    HELLO    ")), rtrim(lit("    HELLO    ")), trim(lit("    HELLO    ")), \
lpad(lit("HELLO"), 10, "*"), rpad(lit("HELLO"), 10, "*")).show(1)
```
