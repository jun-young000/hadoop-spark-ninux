```python
flights2010 = spark.read.optios("data/flights/csv/2010-summary.csv") 
# csv첫줄을 불러오기  #("header", "true")         csv 파일의 가장 윗줄이 컬럼명이고 그것을 제외하고 불러옴을 알수있다.
# 즉 해당 옵션은 첫줄은 헤더라고 옵션을 줬음을 알수 있다. 
flights2010.printSchema()
flights2010.take(5)

# 옵션을 줘서 좀더  정리된 형식으로 읽어오자.
```

```python
# 실행 계획- 꼭 미리 계획하자!!
flights2010.sort("count").explain()
```

```python
# shuffle partition 변경
spark.conf.set("spark.sql.shuffle.partitions", "5")
flights2010.sort("count").take(2)
#partition 은 rdd에서의 파티션 분할개념 5개로 쪼갠다는 소리
# map&reduce 에서 나온거 map에서 나온자료를 reduce로 나눠서 보낼때 섞어서 보냄.최대 200까지 가능 
```

```python
맵리듀스는 맵(Map) 단계와 리듀스(Reduce) 단계로 처리 과정을 나누어 작업
각 단계는 입력과 출력으로써 키-값 쌍을 가지고 있고, 그 타입은 프로그래머가 선택합니다. 또한, 맵과 리듀스 함수도 프로그래머가 직접 작성하게 됩니다

Map은 흩어져 잇는 데이터를 Key, Value의 형태로 연관성 있는 데이터 분류로 묶는 작업( 쉽게 말하면 내가 필요한것을 가져오는 작업) 

Reduce는 Map화한 작업 중 중복 데이터를 제거하고 원하는 데이터를 추출하는 작업
https://over153cm.tistory.com/entry/%EB%A7%B5%EB%A6%AC%EB%93%80%EC%8A%A4MapReduce%EB%9E%80-1
```
