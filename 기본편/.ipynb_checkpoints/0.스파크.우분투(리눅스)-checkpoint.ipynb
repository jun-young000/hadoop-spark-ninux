```python
-rw-rw-r--  test.txt 
	- : directory 아님 (file)
	rw- : 소유자 (생성한 사용자)는 읽기, 쓰기 가능
	rw- : 소유그룹은 읽기, 쓰기 가능
	r-- : 타그룹은 읽기만 가능
```

```python
문제!
test.txt의 타그룹에/ 쓰기 권한 추가 : chmod o+w
test.txt의 소유자에 /실행 권한 추가 : chmod u+x
test.txt의 소유그룹에/ 읽기 권한 삭제 : chmod g-r
```
