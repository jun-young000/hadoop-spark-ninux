```python
spark.sql("""
SELECT LTRIM('    HELLO    '), RTRIM('    HELLO    '), TRIM('    HELLO    '),
LPAD('HELLO', 10, '*'), RPAD('HELLO', 10, '*')
""").show()
```

```python
ltrim(안부)

지정된 문자열 값에 대해 왼쪽 끝에서 공백을 자릅니다.
```

```python
trim(안부)

지정된 문자열 열의 양쪽 끝에서 공백을 자릅니다.
```

```python
rtrim(안부)

지정된 문자열 값에 대해 오른쪽 끝에서 공백을 자릅니다.
```

```python
lpad(col, len, 패드)

len 의 길이만큼 오른쪽 정렬해서 채우고  왼쪽부터 공백 나머지를 패드모양으로 채워라
```

```python
rpad(col, len, 패드)

마우스 오른쪽 패드 문자열 열은 폭 렌 과 패드 .
```

```python
# 정규 표현식~~ 알면  매우 좋은데 이해하기가 매우매우매우 오래걸림.
# 나중이라도 꼬오옥 해보자..!!!
# 로그 처리등에서 정말 도움이 된다...
# 음 다량의 문자열에서 특정 형태 문자열을 뽑아내거나 다른 문자열로 변환 가능
# 특정 패턴의 유효성 검사도 가능...
# 유투브에 좋은 강의 많음
from pyspark.sql.functions import regexp_replace
```

```python
regex_str = "BLACK|WHITE|RED|GREEN|BLUE"
retails.select(regexp_replace(col("DESCRIPTION"), regex_str, "COLOR").alias("color"), \
col("DESCRIPTION")).show(5, False)
```

```python
from pyspark.sql.functions import translate 
# 변환하는건 맞는데 
# 정규식이라는 조금다름 
# translate 은 문자 단위로 치환이 이루어진다고 생각하자.
```

```python
retails.select(translate(col("description"),"ABCD","1234"),col("description")).show(5,False)
# a는 1로 변했고 b는 2로 변했고 c는 3으로 변했고 d는 4로 변했다
```

```python
retails.select(translate(col("description"),"A","1234"),col("description")).show(5,False)
# a는 1로 바뀌고 나머지는 234로 바뀐건 없음.
```

```python
from pyspark.sql.functions import regexp_extract
# 해당 정규식의 패턴을 찾아서~~~ 어쩌구 저쩌구 하자...
extract_str = "(BLACK|WHITE|RED|GREEN|BLUE)"

retails.select(regexp_extract(col("DESCRIPTION"), extract_str, 1).alias("extract"), col("DESCRIPTION")).show(5, False)
```

```python
regexp_extract은  처음 나타난 색상 이름을 추출하는 작업을 수행함..
```

```python
retails.select(regexp_extract(col("DESCRIPTION"), extract_str, 2).alias("extract"), col("DESCRIPTION")).show(5, False)
# 2라고 고치면 어마어마한 에러가.이건 무슨뜻일까??
```

```python
retails.select(regexp_extract(col("DESCRIPTION"), extract_str, 0).alias("extract"), col("DESCRIPTION")).show(5, False)
```

```python
spark.sql("""
SELECT Description FROM retails where instr(Description,'BALCK')>=1 or  instr(Description,'WHITE')>=1
""").show(5)
```

```python
contains_black=instr(col("DESCRIPTION"),"BALCK")>=1
contains_white = instr(col("DESCRIPTION"), "WHITE") >= 1
retails.withColumn("hasBlackWhite", contains_black | contains_white)\
.select("DESCRIPTION", "hasBlackWhite").show(5, False)

# 값 추출 없이 단순히 값의 존재 여부를 확인하고자 할떄.(bool타입)
```

```python
# DESCRIPTION 컬럼에서, BLACK이나 WHITE 라는 단어를 포함한 row만 출력하자. 
# 위에서 true값만 가지는 row만 출력
retails.withColumn("hasBlackWhite", contains_black | contains_white)\
.where("hasBlackWhite").select(col("DESCRIPTION"), col("hasBlackWhite")).show(5, False)
```

```python
#retails.withColumn ("black&white", contains_black|contains_white) 까지만 본다면 참거짓이고
# 여기서 참값만을 가져올려면 .where("hasBlackWhite") 가 필요하다
```
